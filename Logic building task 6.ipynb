```python
# Smart Parking Lot Management System

def parking_system(capacity, logs):

    parked = 0
    peak = 0

    # Count vehicles using logs
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    # Parking status
    if parked > capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

```python
# Online Food Delivery Time Estimator

def delivery_time(distance, traffic, weather):

    # Base time calculation
    time = distance * 5

    # Traffic adjustment
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5

    # Weather adjustment
    if weather == "Rainy":
        time += 5

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
# Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_seats):

    booked = len(booked_seats)

    occupancy = (booked / total_seats) * 100

    # Status check
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)


# Sample Input
booked_seats = [1]*150

seat_occupancy(200, booked_seats)
```

```python
# Cloud Server Load Classification System

def server_load(cpu_readings):

    total = 0

    # Calculate sum
    for value in cpu_readings:
        total += value

    avg = total / len(cpu_readings)

    # Classification
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
# Smart Classroom Resource Usage Monitor

def resource_monitor(resources):

    overused = []

    # Check usage
    for resource in resources:
        if resources[resource] > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

resource_monitor(usage)
```

```python
# Online Event Registration Capacity Controller

def registration_system(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
registration_system(100, 105)
```
