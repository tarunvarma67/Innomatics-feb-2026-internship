```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        amount = int(input("Enter recharge amount: "))
        
        # Rule 1: Minimum recharge amount
        if amount < 50:
            print("Invalid: Minimum recharge amount is ₹50.")
            continue
        
        # Rule 2: Valid plan check
        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid Plan. Please try again.")

# Call the function
# validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Example Data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks_list):
    total = 0
    
    # Calculate total using loop
    for mark in marks_list:
        total += mark
    
    average = total / len(marks_list)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
marks = [60, 45, 70, 55]
result = process_result(marks)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if peak_hour:
        fare += fare * 0.25  # 25% extra
    
    return fare

def cab_fare_system():
    while True:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        peak_hour = True if peak == "yes" else False
        
        total_fare = calculate_fare(distance, peak_hour)
        print("Total Fare: ₹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

# Call the function
# cab_fare_system()
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0
    
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P"]
eligibility = check_attendance(attendance)
print("Eligibility:", eligibility)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"

# Example
password = "Pass@123"
print(check_password(password))
```
